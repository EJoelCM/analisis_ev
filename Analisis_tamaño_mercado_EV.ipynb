```python
# @title
def correccion_electric_range(model,range_mapeo):
  for year, range_value in range_mapeo.items():
    mask = (ev_df['Model'] == model) & (ev_df['Model Year'] == year) & (ev_df['Electric Range'] == 0)
    ev_df.loc[mask, 'Electric Range'] = range_value
```

```python
# @title
year_range_map = {2021:244, 2022:303, 2023:279}
correccion_electric_range('MODEL Y',year_range_map)
year_range_map = {2021:315, 2022:315, 2023:315}
correccion_electric_range('MODEL 3',year_range_map)
year_range_map = {2021:259, 2022:259, 2023:259}
correccion_electric_range('BOLT EV',year_range_map)
year_range_map = {2021:250, 2022:268, 2023:275}
correccion_electric_range('ID.4',year_range_map)
year_range_map = {2021:305, 2022:312, 2023:312}
correccion_electric_range('MUSTANG MACH-E',year_range_map)
year_range_map = {2020:215,2021:226,2022:226,2023:150,2024:212}
correccion_electric_range('LEAF',year_range_map)
year_range_map = {2021: 341, 2022: 351, 2023: 348, 2024: 329}
correccion_electric_range('MODEL X',year_range_map)
correccion_electric_range('IONIQ 5',{2022:303 ,2023:303 ,2024:303})
correccion_electric_range('MODEL S',{2021:303 ,2022:303 ,2023:303})
correccion_electric_range('EV6',{2022:274 ,2023:310 ,2024:310})
ev_df.loc[(ev_df['Electric Range'] == 0) & (ev_df['Model'] == 'KONA ELECTRIC') & (ev_df['Model Year'] == 2021), 'Electric Range'] = 258
```

Para realizar un pronostico del 2024 nos basaremos en el crecimiento que se ha tenido en los años pasados.

Por lo cual se detecta determinamos un crecimiento exonencial por la cantidad de registros, se implementa la fórmula matemática del crecimiento exponencial:

\begin{equation}
y = a e^{bx}
\end{equation}
donde:
*   y = valor en el tiempo x
*   a = amplitud inicial
*   b = tasa de crecimiento
*   e = constante de Euler

```python
# @title
from scipy.optimize import curve_fit

filtrado_anios = ev_conteo_registro[ev_conteo_registro.index <= 2023]
def exp_growth(x, a, b):
    return a * np.exp(b * x)

x_data = filtrado_anios.index - filtrado_anios.index.min()
y_data = filtrado_anios.values
params, covariance = curve_fit(exp_growth, x_data, y_data)
forecast_years = np.arange(2024, 2024 + 6) - filtrado_anios.index.min()
forecasted_values = exp_growth(forecast_years, *params)
forecasted_evs = dict(zip(forecast_years + filtrado_anios.index.min(), forecasted_values))
years = np.arange(filtrado_anios.index.min(), 2029 + 1)
actual_years = filtrado_anios.index
forecast_years_full = np.arange(2024, 2029 + 1)
actual_values = filtrado_anios.values
forecasted_values_full = [forecasted_evs[year] for year in forecast_years_full]
plt.figure(figsize=(12, 8))
plt.plot(actual_years, actual_values, 'go-', label='Registros Actuales')
plt.plot(forecast_years_full, forecasted_values_full, 'ro--', label='Registros Estimados')
plt.title('Mercado EV Actual y Estimado')
plt.xlabel('Año')
plt.ylabel('Numero de EV Registrados')
plt.legend()
plt.grid(True)
plt.show()
```

El pronostico de registros de EVs en los proximos años nos indica que se tendra un aumenmto mas intenso a comparacion de los ultimos años ya que apartir del 2010-2011 se comienza a tener un crecimiento a la alza

Por los datos reales y pronosticados nos indica un mercado en aumento considerable, ya que es probable que esta tendencia continue por los siguientes años, esto podria atraer oportunidades de inversion y negocios relacionados.
